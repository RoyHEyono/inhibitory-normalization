```python
import torch
import torchvision
from inhib_norm.data.mnist import get_sparse_permutation_invariant_fashionmnist_dataloaders
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load FashionMNIST dataset
original_dataloader = get_sparse_permutation_invariant_fashionmnist_dataloaders()
perceptual_invariance_dataloader = get_sparse_permutation_invariant_fashionmnist_dataloaders(brightness_factor=0.75)
```

```python
original_dataiter = iter(original_dataloader['train'])
original_images, _ = next(original_dataiter)
original_images = original_images.reshape(32,1,28,28)
```

```python
pi_dataiter = iter(perceptual_invariance_dataloader['train'])
pi_images, pi_labels = next(pi_dataiter)
pi_images = pi_images.reshape(32,1,28,28)
```

```python
def imshow(img):
    npimg = img.numpy()
    plt.axis('off')
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
imshow(torchvision.utils.make_grid(original_images))
```

```python
imshow(torchvision.utils.make_grid(pi_images))
```
